```python
# Asegurarse de que los nombres coincidan en el tipo de dato
Fusionado['player_name'] = Fusionado['player_name'].astype(str)
Player['display_first_last'] = Player['display_first_last'].astype(str)

# Convertir birthdate a datetime para extraer solo la fecha
Player['birthdate'] = pd.to_datetime(Player['birthdate'])
Player['birthdate'] = Player['birthdate'].dt.date

# Realizar el merge entre Fusionado y Player
Fusionado = Fusionado.merge(
    Player[['display_first_last', 'person_id', 'birthdate']],
    left_on='player_name',
    right_on='display_first_last',
    how='left'
)

# Eliminar la columna 'display_first_last' ya que no es necesaria en el resultado final
Fusionado = Fusionado.drop(columns=['display_first_last'])

# Mostrar el DataFrame actualizado
print(Fusionado.head())
```

```python
# Crear una copia para trabajar y evitar efectos colaterales
Fusionado_filled = Fusionado.copy()

# Iterar por cada columna que queremos llenar
for columna in ['person_id', 'birthdate']:
    # Crear un mapeo único por player_name para no repetirse
    mapeo = Fusionado_filled.dropna(subset=[columna]).drop_duplicates(subset=['player_name'])[['player_name', columna]]
    mapeo_dict = dict(zip(mapeo['player_name'], mapeo[columna]))
    
    # Usar el mapeo para llenar los valores nulos
    Fusionado_filled[columna] = Fusionado_filled.apply(
        lambda row: mapeo_dict.get(row['player_name'], row[columna]) if pd.isnull(row[columna]) else row[columna],
        axis=1
    )

# Mostrar un ejemplo del DataFrame resultante
print(Fusionado_filled.head())
```

```python

# Copiar el DataFrame original para trabajar
Fusionado_actualizado = Fusionado.copy()

# Modificar la columna person_id para eliminar el ".0"
Fusionado_actualizado['person_id'] = Fusionado_actualizado['person_id'].astype(str).str.replace('.0', '', regex=False)

# Asegurar que las columnas game_date y birthdate sean del tipo datetime
Fusionado_actualizado['game_date'] = pd.to_datetime(Fusionado_actualizado['game_date'])
Fusionado_actualizado['birthdate'] = pd.to_datetime(Fusionado_actualizado['birthdate'])

# Calcular la edad del jugador en el momento del partido
Fusionado_actualizado['edad'] = Fusionado_actualizado.apply(
    lambda row: row['game_date'].year - row['birthdate'].year - 
    ((row['game_date'].month, row['game_date'].day) < (row['birthdate'].month, row['birthdate'].day)),
    axis=1
)

# Mostrar un ejemplo del DataFrame actualizado
print(Fusionado_actualizado[['player_name', 'game_date', 'birthdate', 'edad', 'person_id']].head())
```

```python
# Convertir la columna 'game_date' a formato datetime si no lo está
Fusionado_actualizado['game_date'] = pd.to_datetime(Fusionado_actualizado['game_date'])

# Función para calcular la temporada
def determinar_temporada(fecha):
    año = fecha.year
    if fecha.month >= 9:  # Si el mes es septiembre (9) o después, es el inicio de la nueva temporada
        return f"{año}-{año + 1}"
    else:  # Si el mes es antes de septiembre, pertenece a la temporada del año anterior
        return f"{año - 1}-{año}"

# Aplicar la función al DataFrame
Fusionado_actualizado['season'] = Fusionado_actualizado['game_date'].apply(determinar_temporada)

Fusionado_actualizado
```

```python
import pandas as pd
import numpy as np

# Excluir columnas que no se agruparán
columnas_a_sumar = [
    "Total_Tiros_Libres_IN", "Total_Tiros_Libres_OUT", "Block_Count",
    "Missed_Shots_Count", "Max_Assists", "Max_Points", "Steals",
    "Lost Ball", "Bad Pass", "Foul", "Max_Total_Rebounds", "Total_puntos"
]

# Calcular edad promedio por jugador y temporada
def calcular_edad_promedio(edad_group):
    return np.mean(edad_group)

# Agrupar y sumar las columnas, excepto la edad
Fusionado_season = (
    Fusionado_actualizado.groupby(["season", "player_name"])
    .agg({
        **{col: "sum" for col in columnas_a_sumar},  # Sumar las columnas
        "edad": calcular_edad_promedio               # Calcular la edad promedio
    })
    .reset_index()  # Restablecer el índice para mantener la estructura
)

# Reorganizar las columnas: primero 'season', luego 'player_name'
Fusionado_season = Fusionado_season[["season", "player_name"] + columnas_a_sumar + ["edad"]+ ["game_date"]]

# Mostrar un vistazo del DataFrame resultante
print(Fusionado_season.head())



```

MODELO DE MACHINE LEARNING

El objetivo es construir un modelo de clasificación que identifique si un jugador entre 19 y 23 años es un talento prometedor, basándonos en estadísticas de partidos.

Definición de "Talento"
Se establecen reglas para definir qué es un "talento".

Un jugador es un talento si supera ciertos umbrales basados en la temporada 2004-2005 de LeBron James a la edad de 19

```python
# Importar librerías
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score



# Crear la columna de 'talento' (simplificada)
Fusionado['talento'] = (Fusionado['Total_puntos'] > 800).astype(int)

# Dividir los datos en entrenamiento y prueba
entrenamiento = Fusionado[Fusionado['season'] <= '2019-2020']
prueba = Fusionado[(Fusionado['season'] == '2022-2023') & (Fusionado['edad'] < 22)]

# Seleccionar las características y la etiqueta
features = ['Total_Tiros_Libres_IN', 'Max_Assists', 'Max_Points', 'Steals', 'Block_Count']
X_train = entrenamiento[features]
y_train = entrenamiento['talento']
X_test = prueba[features]

# Modelo: Árbol de Decisión
modelo = DecisionTreeClassifier(max_depth=3, random_state=42)
modelo.fit(X_train, y_train)

# Predicciones
y_pred = modelo.predict(X_test)

# Mostrar los resultados
prueba['Predicción Talento'] = y_pred
print("Jugadores Predichos como Talento en 2022-2023:")
print(prueba[['player_name', 'Predicción Talento']])
```

```python
# Filtrar jugadores predichos como talento
jugadores_talento = prueba[prueba['Predicción Talento'] == 1]

# Seleccionar las columnas deseadas
jugadores_talento = jugadores_talento[['player_name', 'edad', 'Total_puntos', 'Max_Assists', 'Block_Count', 'Max_Total_Rebounds', 'Steals']]

# Eliminar jugadores con edad = 0
jugadores_talento = jugadores_talento[jugadores_talento['edad'] > 0]

# Convertir los valores numéricos a enteros
columnas_a_convertir = ['edad', 'Total_puntos', 'Max_Assists', 'Block_Count', 'Max_Total_Rebounds', 'Steals']
jugadores_talento[columnas_a_convertir] = jugadores_talento[columnas_a_convertir].astype(int)

# Mostrar el DataFrame resultante
print("Jugadores Predichos como Talento:")
jugadores_talento


```

El modelo **DecisionTreeClassifier** es un algoritmo de aprendizaje supervisado que pertenece a la familia de los árboles de decisión. Su propósito principal es resolver problemas de clasificación (aunque tiene una variante para regresión). A continuación, funcionamiento:

---

### 1. **Estructura básica**
Un árbol de decisión se construye como una estructura jerárquica similar a un diagrama de flujo, donde:
- **Nodos internos** representan una decisión basada en una característica específica de los datos.
- **Ramas** representan los posibles resultados de una decisión.
- **Hojas** representan las categorías o clases finales (etiquetas predichas).

---

### 2. **Proceso de construcción**
El modelo se construye dividiendo repetidamente los datos en subconjuntos más pequeños con base en una característica, siguiendo estos pasos:

#### **a. Selección de características clave**
El algoritmo selecciona la característica que mejor separa las clases de los datos. Para medir esta "mejor separación", se utilizan métricas como:
- **Gini Impurity**: Mide cuán "puras" son las divisiones.
- **Entropy/Information Gain**: Cuantifica la ganancia de información al dividir un nodo.

#### **b. División recursiva**
- Una vez elegida la característica más relevante, los datos se dividen en subconjuntos (ramas) según los valores de esa característica.
- Este proceso se repite para cada subconjunto, creando nodos y ramas adicionales.

#### **c. Condición de parada**
El árbol no crece indefinidamente; se detiene según criterios predefinidos, como:
- Una profundidad máxima.
- Un número mínimo de muestras en un nodo.
- Si las divisiones no mejoran significativamente la clasificación.

---

### 3. **Predicción**
Para clasificar un nuevo dato:
1. Se inicia en la raíz del árbol.
2. Se evalúan las condiciones en los nodos para determinar qué rama seguir.
3. El proceso continúa hasta llegar a una hoja, donde se asigna la clase predicha.

---

### 4. **Ventajas**
- **Simplicidad y claridad**: Fácil de interpretar y visualizar.
- **Poca preparación de datos**: No requiere escalado de características ni normalización.
- **Manejo de datos no lineales**: Puede capturar relaciones no lineales en los datos.
- **Capacidad de manejar características categóricas**: A diferencia de otros modelos, no requiere codificación especial.

---

### 5. **Desventajas**
- **Propenso al sobreajuste**: Si no se controla su crecimiento (sin podar), puede aprender patrones específicos del conjunto de entrenamiento y perder generalización.
- **Sensibilidad a los datos**: Cambios pequeños en los datos pueden producir árboles muy diferentes.
- **Menor rendimiento en conjuntos grandes**: En comparación con modelos más sofisticados como Random Forests o Gradient Boosting.
